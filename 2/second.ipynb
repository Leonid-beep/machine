## Предметная область, источник и характер данных
Был выбран датасет, который содержит в себе 2000 строк данных о домах с различными характеристиками, которые могут повлиять на цены на жилье. Данные являются реальными - рынки недвижимости, такие как рынки Сиднея и Мельбурна, предоставляют аналитикам данных интересную возможность для анализа и прогнозирования динамики цен на недвижимость. Датасет был взят с сайта https://www.kaggle.com/datasets/zafarali27/house-price-prediction-dataset
## Добавление нового атрибута
Добавим атрибут под названием "средняя площадь комнаты (Avg Room Area)", рассчитываемый по формуле: Area/(Bedrooms+Bathrooms). Выведем нашу таблицу, а затем добавим в нее новый атрибут:

```python
import numpy as np
import pandas as pd
import scipy
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import figure
%matplotlib inline

df = pd.read_csv("./house-price.csv")
df.head()
```

```python
list = []
for i, row in df.iterrows():
    area = row['Area']
    bedrooms = row['Bedrooms']
    bathrooms = row['Bathrooms']
    avgRoomArea = area/(bedrooms+bathrooms)
    list.append(avgRoomArea)

avgRoomArea = pd.Series(list)
df.insert(0, "avgRoomArea", avgRoomArea)
```

```python
df.info()
df.head()
```

Как мы видим, атрибут был добавлен в датасет.

## "Причёсывание датасета"
Удалим ненужные данные, например, текстовые и ID:

```python
df = df.drop(['Id', 'Location', 'Condition', 'Garage'], axis="columns")
df = df.dropna()
```

```python
df.info()
```

Найдем выбросы и удалим их из датасета:

```python
for col in df.columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers_count = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col].count()
    print(f"Колонка {col} - {outliers_count} выбросов")

df_filtered = df.copy()
for col in df_filtered.columns:
    Q1 = df_filtered[col].quantile(0.25)
    Q3 = df_filtered[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_filtered = df_filtered[(df_filtered[col] >= lower_bound) & (df_filtered[col] <= upper_bound)]

df_filtered.info()
```

## Построение графиков
Построим графики зависимости одной переменной от другой:

```python
sns.pairplot(df_filtered)
plt.show()
```

#### avgRoomArea и Area: 

Точки выстраиваются в диагональные кластеры, между этими параметрами есть положительная корреляция.

#### Area и Price:

Также наблюдается положительная корреляция, с увеличением площади чаще всего увеличивается и цена.

#### Price и остальные параметры:

На графиках также видно, что цена слегка выше у объектов с большим числом спален и ванных комнат.

## Применение метода KMeans

```python
df_filtered.head()
```

Сейчас так выглядит наш датасет. Применим к нему метод KMeans

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_filtered)
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.xlabel('Количество кластеров')
plt.ylabel('WCSS')
plt.show()
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(scaled_data)
df_filtered['Cluster'] = kmeans.labels_
```

## Визуализация кластеров
Визуализируем кластеры по всем атрибутам с помощью PCA:

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])

pca_df['Cluster'] = df_filtered['Cluster']

plt.figure(figsize=(10, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=pca_df, palette='viridis', s=100)
plt.title('PCA of House Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(loc='best')
plt.grid()
plt.show()
```
