# Лабораторная работа №3
## Предметная область, источник и характер данных
Был выбран датасет, который содержит в себе около 4000 строк данных об абалонах (морские ракушки). Датасет рассчитан на определение возраста абалона по его характеристикам. Датасет был взят с сайта https://www.kaggle.com/datasets/devzohaib/estimating-the-age-of-abalone-at-a-seafood-farm
## Описание атрибутов
- Пол (sex): пол абалона может быть мужским, женским, а также неопределенным (детским), так как пол абалона можно определить только в определенном возрасте.
- Длина (length): самое длинное измерение длины ракушки моллюска.
- Диаметр (diameter): перпендикуляр к длине абалона.
- Высота (height): высота ракушки.
- Целый вес (whole_wt): целый вес абалона.
- Внутренний вес (shucked_wt): вес внутренностей абалона.
- Вес кишечника (gut_wt): вес кишечника абалона.
- Вес ракушки (shell_wt): вес самой ракушки абалона.
- Кольца (rings): количество колец на ракушке.
- Возраст (age): возраст абалона.
## Описание задачи анализа
Исследование набора данных с целью выполнения классификации. Будем предсказывать пол абалона на основе всех данных
## Работа с датасетом
Выведем датасет:

```python
import numpy as np
import pandas as pd
import scipy
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import figure
%matplotlib inline

df = pd.read_csv("./abalone.csv")
df.head()
```

Удалим выбросы и пустые значения:

```python
df_filled = df.copy()
numeric_cols = df.select_dtypes(include=[np.number]).columns
df_filled[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
Q1 = df_filled[numeric_cols].quantile(0.25)
Q3 = df_filled[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = df_filled[~((df_filled[numeric_cols] < lower_bound) | (df_filled[numeric_cols] > upper_bound)).any(axis=1)]

df_cleaned = df_no_outliers.dropna()
print("До обработки:", df.shape)
print("После удаления выбросов:", df_no_outliers.shape)
print("После удаления пустых значений:", df_cleaned.shape)

df_cleaned.head()
```

## Классификация
В качестве целевого класса выберем пол абалона (sex).
Проверим сбалансированность классов:

```python
class_distribution = df_cleaned['sex'].value_counts()
print(class_distribution)
```

Как видим, данные не сбалансированы. Займемся балансировкой данных:

```python
from imblearn.over_sampling import RandomOverSampler

X = df_cleaned.drop(columns='sex')
y = df_cleaned['sex']

ros = RandomOverSampler(random_state=42)

X_resampled, y_resampled = ros.fit_resample(X, y)

df_balanced = pd.concat([X_resampled, y_resampled], axis=1)

class_distribution_balanced = df_balanced['sex'].value_counts()
print(class_distribution_balanced)
```

Теперь данные сбалансированы. Далее поделим данные на тренировочные и тестовые:

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = df_balanced.drop(columns='sex')
Y = df_balanced['sex']
encoder = LabelEncoder()
Y = encoder.fit_transform(Y)
scaler = StandardScaler()
X = scaler.fit_transform(X)
X_tr, X_t, Y_tr, Y_t = train_test_split(X, Y, stratify=Y)
```

## kNN
Найдем оптимальный параметр k и произведем саму классификацию:

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn .tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import roc_curve, auc
from yellowbrick.classifier import ClassificationReport
from yellowbrick.classifier import ROCAUC

knn = KNeighborsClassifier()
grid_params = {'n_neighbors': [i for i in range(1, 31)], 'weights': ['uniform', 'distance'], 'metric': ['minkowski', 'euclidean', 'manhattan']}
gs = GridSearchCV(knn, grid_params, scoring='accuracy', refit=True)
g_res = gs.fit(X_tr, Y_tr)
k = g_res.best_params_['n_neighbors']
metric = g_res.best_params_['metric']
weights = g_res.best_params_['weights']
g_res.best_params_
```

```python
knn = KNeighborsClassifier(n_neighbors=k, weights=weights).fit(X_tr, Y_tr)
predicted = knn.predict(X_t)
```

## Оценка качества классификации с помощью метода knn
### Accuracy-оценка

```python
r = 0
w = 0
for i, pos in enumerate(Y_t):
    if predicted[i] == pos:
        r += 1
    else:
        w += 1
accuracy = r / (r + w)
print(accuracy)
print(f"{accuracy * 100:.2f}%")
```

### Presicion, Recall, F-measure-оценка

```python
report = ClassificationReport(knn, classes=encoder.classes_)
report.fit(X_tr, Y_tr)
report.score(X_t, Y_t)
report.show();
```

### ROC-оценка

```python
report = ROCAUC(knn, classes=encoder.classes_)
report.fit(X_tr, Y_tr)
report.score(X_t, Y_t)
report.show();
```

## Дерево решений

```python
dt = DecisionTreeClassifier()
dt.fit(X_tr, Y_tr)
dt_predicted = dt.predict(X_t)
```

## Оценка качества классификации с помощью метода дерева решений
### Accuracy-оценка

```python
r = 0
w = 0
for i, pos in enumerate(Y_t):
    if dt_predicted[i] == pos:
        r += 1
    else:
        w += 1

accuracy = r / (r + w)
print(accuracy)
print(f"{accuracy * 100:.2f}%")
```

### Presicion, Recall, F-measure-оценка

```python
report = ClassificationReport(dt, classes=encoder.classes_)
report.fit(X_tr, Y_tr)
report.score(X_t, Y_t)
report.show();
```

### ROC-оценка

```python
report = ROCAUC(dt, classes=encoder.classes_)
report.fit(X_tr, Y_tr)
report.score(X_t, Y_t)
report.show();
```

## Вывод

В ходе выполнения лабораторной работы был выбран новый датасет для проведения классификации: для этого мы выбрали целевой класс в нашем датасете, затем с помощью методов kNN и дерева решений провели классификацию. Затем мы оценили качество классификации каждого из методов с помощью таких методов как Accuracy, Presicion, Recall, F-measure и ROC. Можно заметить, что разница между классификацией методами kNN и деревом решений небольшая. Теперь рассмотрим подробнее каждый класс:

1) Наиболее легким для предсказания является "детский" пол. При обеих классификациях он показал наивысшие результаты. Например, при ROC-оценке AUC в методе kNN = 0.91, а при методе дерева решений AUC = 0.78, что выше оставшихся значений.
2) Мужской и женский пол абалона в обоих методах показывает среднюю точность, в обеих методах при ROC-оценке их AUC колеблется районе 0.57-0.75, а при других оценках является очень маленьким: например, точность мужского пола при kNN-классификации при recall-оценке равна 0.344.
